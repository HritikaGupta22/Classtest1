```python
# Creating a Dictionary
# Update the Dictionary by adding columns
# convert Dictionary into DataFrame
# Update DataFrame by adding Columns and Rows
# EXtract Data in a required Format
# Persform some operation on the data
```

```python
# Create a Employee Dictionary 'emp_dict' with following key as 'Name', 'Education' and 'Gender'
emp_dict = {
    'Name': ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQR', 'STU'],
    'Education': ['Graduate', 'Postgraduate', 'Postgraduate', 'Graduate', 'Graduate', 'Postgraduate', 'Graduate'],
    'Gender': ['Male', 'Female', 'Male', 'Other', 'Female', 'Female', 'Male']
}
print(emp_dict)
```

```python
#Update the Employee Dictionary 'emp_dict' with following Information :
# Add new Column 'Age'
emp_dict['Age']=[22,27,26,23,24,30,21]
print(emp_dict)
```

```python
#Create a Dataframe 'emp_df' from the Dictionary 'emp_dict'
#importing pandas
import pandas as pd
#Pandas function DataFrame will convert dictonary into data frame
emp_df = pd.DataFrame(emp_dict)
print(emp_df)
```

```python
# Update the Dataframe 'emp_df' with the following Information:
# add columns 'Salary_Lakhs' and 'Bonus%'

#Define the list of salary values in lakhs for each employee
salary_lakhs = [6, 15, 20, 5, 10, 18, 12]
#Define the list of Bonus in ercentage for each employee
bonus_percent = [12.50, 8.75, 6.25, 10.20, 13.60, 11.40, 9.80]
# Update the 'emp_df' DataFrame by adding a new column 'Salary_Lakhs' with salary values
emp_df['Salary_Lakhs'] = salary_lakhs
# Update the 'emp_df' DataFrame by adding a new column 'Bonus%' with bonus_percent values
emp_df['Bonus%'] = bonus_percent
# print the update data frame
print(emp_df)
```

```python
# Update the Dataframe ‘emp_df’ with the following information:
# Adding 3 new rows to the existing table

# Define a dictionary containing the new data to be added
new_data = {
    'Name': ['VWX', 'YZA', 'BCD'],
    'Education': ['Postgraduate', 'Graduate', 'Postgraduate'],
    'Gender': ['Male', 'Female', 'Other'],
    'Age': [35, 28, 32],
    'Salary_Lakhs': [14, 7, 8],
    'Bonus%': [5.50, 7.75, 14.80]
}
# Create a DataFrame using the new data
new_rows_df = pd.DataFrame(new_data)
# Append the new rows to the existing 'emp_df' DataFrame
emp_df = pd.concat([emp_df,new_rows_df])
#print the updated emp_df
print(emp_df)
```

```python
# Create a Column ‘Gross_Salary_Lakhs’ with the following information:
# Gross_Salary_Lakhs = Salary_Lakhs * (1 + Bonus%)
emp_df['Gross_Salary_Lakhs'] = emp_df['Salary_Lakhs'] * (1 + emp_df['Bonus%'] / 100)
# Print the updated 'emp_df'
print(emp_df)
```

```python
# Create a Subset ‘emp_df_ss’ from ‘emp_df’ with the following Variables: {Name, Age, Gross_Salary_Lakhs}
# List of columns to include in the subset
selected_columns = ['Name', 'Age', 'Gross_Salary_Lakhs']

# Create the subset DataFrame emp_df_ss
emp_df_ss = emp_df[selected_columns]
emp_df_ss
```

```python
# Create a Copy of 'emp_df' Named as 'emp_df_age_sorted' and Sort {Highest to Lowest} by 'Age'
emp_df_age_sorted = emp_df.copy()

# Sort emp_df_age_sorted by 'Age' in descending order
emp_df_age_sorted = emp_df_age_sorted.sort_values(by='Age', ascending=False)

#print the sorted data
print(emp_df_age_sorted)
```

```python
# Create a Copy of 'emp_df' Named as 'emp_df_age_salary_sorted' and Sort: First by 'Age' {Lowest to Highest}, Second by 'Gross_Salary_Lakhs' {Highest to Lowest}
# Create a copy of emp_df
emp_df_age_salary_sorted = emp_df.copy()
# Sort emp_df_age_salary_sorted first by 'Age' (ascending), then by 'Gross_Salary_Lakhs' (descending)
emp_df_age_salary_sorted = emp_df_age_salary_sorted.sort_values(by=['Age', 'Gross_Salary_Lakhs'], ascending=[True, False])

#print the data sorted by age and salary
print(emp_df_age_salary_sorted)
```

```python
# Create a Dataframe ‘emp_df_filtered’ to Filter ‘emp_df’ using the following Information: 'Age' >= 25 & 'Gender' = 'Female'
# Creating 'emp_df_filtered' by applying the filter
emp_df_filtered = emp_df[(emp_df['Age'] >= 25) & (emp_df['Gender'] == 'Female')]
# print the filtered data frame
print(emp_df_filtered)
```

```python
# Create 2 Subsets: ‘emp_df_grad’ & ‘emp_df_postgrad’ from ‘emp_df’ containing Information of Employees having ‘Education’ as ‘Graduate’ & ‘Postgraduate’, respectively
# Create 'emp_df_grad' containing employees with 'Education' as 'Graduate'
emp_df_grad = emp_df[emp_df['Education'] == 'Graduate']
# Create 'emp_df_postgrad' containing employees with 'Education' as 'Postgraduate'
emp_df_postgrad = emp_df[emp_df['Education'] == 'Postgraduate']
# print the data frames containing Graduation and Postgraduation as Education Respectively
print("Data frame containing only Graduate\n",emp_df_grad)
print("Data frame containing only Postgraduate\n",emp_df_postgrad)
```

```python
# Create a Dataframe ‘emp_df_merged’ to Inner Merge ‘emp_df_grad’ having only following Variables {Name, Gender, Age} with ‘emp_df_postgrad’ having only following Variables {Name, Gender, Gross_Salary_Lakhs} on ‘Gender
# Creating emp_df_grad and emp_df_postgrad are our DataFrames with column names as required
# Creating emp_df_grad’ having only following Variables {Name, Gender, Age}
columns_grad = ['Name', 'Gender', 'Age']
emp_df_grad = emp_df[columns_grad]

# Creating emp_df_postgrad’ having only following Variables {Name, Gender, Gross_Salary_Lakhs}
columns_postgrad = ['Name', 'Gender', 'Gross_Salary_Lakhs']
emp_df_postgrad = emp_df[columns_postgrad]
# Perform an inner merge on 'Gender' to create emp_df_merged
emp_df_merged = emp_df_grad.merge(emp_df_postgrad, on='Gender', how='inner')
# Print the merged data frame
print(emp_df_merged)
```

```python
# Group ‘emp_df’ to Create a Table ‘emp_df_gen_edu’ using ‘Gender’ & ‘Education’ having Count of Employees
# emp_df is original DataFrame
# Grouping by 'Gender' and 'Education' and counting employees in each group and index count of employees as COUNT_E
emp_df_gen_edu = emp_df.groupby(['Gender', 'Education']).size().reset_index(name='COUNT_E')
# print the data frame emp_df_gen_edu
(emp_df_gen_edu)
```

```python
# Group ‘emp_df’ to Create a Table ‘emp_df_gen_age_sal’ using ‘Gender’ with Average of Variables {Age & Gross_Salary_Lakhs}
# emp_df is our original DataFrame
# Grouping by 'Gender' and calculating the average of 'Age' and 'Gross_Salary_Lakhs' in each group
emp_df_gen_age_sal = emp_df.groupby('Gender')[['Age', 'Gross_Salary_Lakhs']].mean().reset_index()
# Print the data frame emp_df_gen_age_sal
print(emp_df_gen_age_sal)
```

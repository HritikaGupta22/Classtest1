```python
# 1.1. Arrays : Dimensions 



import numpy as np

# 1-Dimensional Array
a1 = np.array([1, 2, 3, 4], dtype=float)
print(a1)
print()
print(type(a1))
print()
print(a1.shape)  # Output: (4,) - Dimension = 4 Rows, 1 Column
print()
# 2-Dimensional Array
a2 = np.array([[11, 12, 13], [21, 22, 23]])
print(a2)
print()
print(type(a2))
print()
print(a2.shape)  # Output: (2, 3) - Dimension = 2 Rows, 3 Columns
print()
# 3-Dimensional Array
a3 = np.array([[[11, 12, 13], [21, 22, 23]], [[31, 32, 33], [41, 42, 44]]])
print(a3)
print()
print(type(a3))
print()
print(a3.shape)  # Output: (2, 2, 3) - Dimension = 2 Layers, 2 Rows, 3 Columns

```

```python
# 1.2. Arrays : Manipulation



import numpy as np

a1 = np.array([1, 2, 3, 4])


#1-d Array Indexing
print("a1: ",a1[0])  # Output: 1 - Extract First Element

print ()
#1-d Array Reshape >> 2-d Array {4 Rows, 1 Column >> 2 Rows, 2 Columns}
print("after reshaping : ",a1.reshape(2, 2))
a2 = np.array([[11, 12, 13], [21, 22, 23]])

print ()
#2-d Array Indexing
print(a2[0])  # Output: [11 12 13] - Extract First Row Elements
print ()
print(a2[1, 2])  # Output: 23 - Extract Second Row Third Element

print ()
#2-d Array Reshape >> 2-d Array {2 Rows, 3 Columns >> 3 Rows, 2 Columns}
print(a2.reshape(3, 2))


a3 = np.array([[[11, 12, 13], [21, 22, 23]], [[31, 32, 33], [41, 42, 44]]])
print("a3",a3)
print ()
#3-d Array Indexing
print(a3[0])

print ()
#3-d Array Reshape >> {2 Layers, 3 Rows, 2 Columns}
print("after reshaping : ",a3.reshape(2, 3, 2))

print ()
#Array Reshape >> 3-d --> 2-d Array
#print("after reshaping unton 2d array ",a3.reshape(3, 4))

print ()
#2-d Array Transpose : (2, 3) --> (3, 2)
print("transpose of a2 : ",a2.T)

print ()

#3-d Array Transpose : (2, 3, 2) --> (2, 2, 3)
print("transpose of a3 : ",a3.transpose())
print("transpose with dimension : ")
print(np.transpose(a3, axes=(0, 2, 1)))
```

```python
# 1.4 Arrays : Statistics



import numpy as np

# Create arrays
a4 = np.array([1, 23, 4])
a5 = np.array([56, 7, 8])

# Concatenate & Reshape
a45 = np.concatenate((a4, a5)).reshape(2, 3)

# Array Min
min_val = np.amin(a45)
min_col = np.amin(a45, axis=0)  # Column Min over Row
min_row = np.amin(a45, axis=1)  # Row Min over Column

# Array Max
max_val = np.amax(a45)
max_col = np.amax(a45, axis=0)  # Column Max over Row
max_row = np.amax(a45, axis=1)  # Row Max over Column

# Array Sum
sum_val = np.sum(a45)
sum_col = np.sum(a45, axis=0)  # Column Sum over Row
sum_row = np.sum(a45, axis=1)  # Row Sum over Column

# Array Mean
mean_val = np.mean(a45)
mean_col = np.mean(a45, axis=0)  # Column Mean over Row
mean_row = np.mean(a45, axis=1)  # Row Mean over Column

# Array Std.Dev.
std_dev_val = np.std(a45)
std_dev_col = np.std(a45, axis=0)  # Column Std.Dev. over Row
std_dev_row = np.std(a45, axis=1)  # Row Std.Dev. over Column

# Array Covariance & Correlation
covariance = np.cov(a4, a5)  # Covariance
correlation = np.corrcoef(a4, a5)  # Pearson Correlation

# Print results
print("a4 =", a4)
print()
print("a5 =", a5)
print()
print("a45 =", a45)
print()
print("Minimum value of a45:", min_val)
print()
print("Column-wise minimum of a45:", min_col)
print()
print("Row-wise minimum of a45:", min_row)
print()
print("Maximum value of a45:", max_val)
print()
print("Column-wise maximum of a45:", max_col)
print()
print("Row-wise maximum of a45:", max_row)
print()
print("Sum of a45:", sum_val)
print()
print("Column-wise sum of a45:", sum_col)
print()
print("Row-wise sum of a45:", sum_row)
print()
print("Mean of a45:", mean_val)
print()
print("Column-wise mean of a45:", mean_col)
print()
print("Row-wise mean of a45:", mean_row)
print()
print("Standard deviation of a45:", std_dev_val)
print()
print("Column-wise standard deviation of a45:", std_dev_col)
print()
print("Row-wise standard deviation of a45:", std_dev_row)
print()
print("Covariance between a4 and a5:", covariance)
print()
print("Correlation between a4 and a5:", correlation)
```

```python
#1.5 Numpy Application : Linear Algebra




import numpy as np

# Define matrices a6 and a7
a6 = np.array([[1, 3], [2, 4]])  # Matrix a6
a7 = np.array([[5, 7], [6, 8]])  # Matrix a7

# Vector Multiplication: {1*5 + 3*7 + 2*6 + 4*8}
a67_vm = np.vdot(a6, a7)  # Vector multiplication of a6 and a7

# Matrix Multiplication: {1*5 + 3*6, 1*7 + 3*8, 2*5 + 4*6, 2*7 + 4*8}
a67_mm1 = np.dot(a6, a7)  # Matrix multiplication of a6 and a7 using np.dot
a67_mm2 = np.matmul(a6, a7)  # Matrix multiplication of a6 and a7 using np.matmul

# Matrix Multiplication
a67 = np.dot(a6, a7)  # Matrix multiplication of a6 and a7

# Matrix Determinant
a67_det = np.linalg.det(a67)  # Determinant of matrix a67

# Matrix Inverse
a67_inv = np.linalg.inv(a67)  # Inverse of matrix a67

# Matrix Eigen Value & Eigen Vector
eig_val, eig_vec = np.linalg.eig(a67)  # Eigenvalues and eigenvectors of matrix a67

# System of Linear Equations: Solution
a8 = np.array([[1, 2], [3, 2]])  # Coefficient matrix
a9 = np.array([4, 8])  # Right-hand side vector

# Solve for x and y
solution = np.linalg.solve(a8, a9)  # Solution of the system of linear equations

# Print the results
print("a6:\n", a6)
print()
print("a7:\n", a7)
print()
print("a67_vm:", a67_vm)          # Vector multiplication of a6 and a7
print()
print("a67_mm1:\n", a67_mm1)       # Matrix multiplication of a6 and a7 using np.dot
print()
print("a67_mm2:\n", a67_mm2)      # Matrix multiplication of a6 and a7 using np.matmul
print()
print("a67:\n", a67)           # Matrix multiplication of a6 and a7
print()
print("a67_det:", a67_det)     # Determinant of matrix a6
print()
print("a67_inv:\n", a67_inv)    # Inverse of matrix a67
print()
print("eig_val:", eig_val)      # Eigenvalues and eigenvectors of matrix a67
print()
print("eig_vec:\n", eig_vec)    # Eigenvalues and eigenvectors of matrix a67
print()
print("Solution:", solution)  # Solution of the system of linear equations
```

```python
# 2.1. Pandas Series (with Index)



import pandas as pd  # Importing the pandas library

# Series with Natural Index: {0, 1, 2, ...}
my_series1 = ['a', 'b', 'c', 'd']  # Creating a list of values
ps1 = pd.Series(my_series1)  # Creating a pandas Series using the list
# The Series will have default integer index starting from 0
# ps1 will contain: 
# 0    a
# 1    b
# 2    c
# 3    d
# dtype: object

# Series with Given Index: {'a', 'b', 'c', ...}
my_series2 = {'a': 12, 'b': 34, 'c': 56}  # Creating a dictionary with key-value pairs
ps2 = pd.Series(my_series2)  # Creating a pandas Series using the dictionary
# The Series will have the given keys as index and corresponding values
# ps2 will contain:
# a    12
# b    34
# c    56
# dtype: int64
print("series 1:",ps1)
print()
print("Series 2:",ps2)
print()
print("print 2nd value of Series 2:",ps2[1])
print()
print("Print Value using variable:",ps2['a'])  # Prints the value 12 using the index 'a'
print()
print("print Variable using Value:",ps2[ps2 == 12])  # Prints the index 'a' using the value 12

```

```python
#pivot table 



import pandas as pd

# Create DataFrame: Using Python Dict
my_data = {
    'Name': ['Alexa', 'Bixby', 'Cortana', 'Jarvis', 'Lyra', 'Robin', 'Siri'],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Education': ['Postgraduate', 'Graduate', 'Postgraduate', 'Doctorate', 'Graduate', 'Graduate', 'Postgraduate'],
    'Age': [24, 21, 25, 28, 23, 22, 24],
    'Monthly_Salary': [27000, 25000, 30000, 35000, 26000, 24000, 28000]
}

# Create a DataFrame from the my_data dictionary
my_data_df = pd.DataFrame(my_data)

# Print the original DataFrame
print("Original 'my_data_df' DataFrame:")
print(my_data_df)

# Pivot DataFrame
my_data_df_pivot = my_data_df.pivot_table(index=['Education', 'Gender'], columns='Name', values=['Age', 'Monthly_Salary'])

# Reset index of the pivot table
my_data_df_pivot.reset_index(inplace=True)

# Print the pivoted DataFrame
print("\nPivoted 'my_data_df_pivot' DataFrame:")
print(my_data_df_pivot)
```

```python
#Meld function



import pandas as pd
# Create Dummy DataFrame: Using Python Dict
panel_data = {
    'Company': ['ABC', 'ABC', 'DEF', 'DEF', 'GHI', 'GHI'],
    'Year': [2001, 2002, 2001, 2002, 2001, 2002],
    'Revenue': [1234, 2345, 3456, 4567, 5678, 6789]
}

# Create a DataFrame from the panel_data dictionary
panel_df = pd.DataFrame(panel_data)

# Pivot DataFrame
panel_df_pivot = pd.pivot(panel_df, index='Company', columns='Year', values='Revenue')  # Create a pivot table from the DataFrame
panel_df_pivot.reset_index(level='Company', inplace=True)  # Reset the index of the pivot table

# Print the pivoted DataFrame
print("\nPivoted 'panel_df_pivot' DataFrame:")
print(panel_df_pivot)
print()

# Melt DataFrame
panel_df_pivot_melt = pd.melt(panel_df_pivot, id_vars=['Company'], value_name='Revenue')  # Convert the pivot table to a long format

# Print the melted DataFrame
print("Melted 'panel_df_pivot_melt' DataFrame:")
print(panel_df_pivot_melt)

# Sort the melted DataFrame by 'Revenue' column
panel_df_pivot_melt_sorted = panel_df_pivot_melt.sort_values(by=['Revenue'])

# Print the sorted melted DataFrame
print("Sorted 'panel_df_pivot_melt_sorted' DataFrame:")
print(panel_df_pivot_melt_sorted)
```

```python
# 2.4. Combine Dataframe : Concat, Merge



import pandas as pd

# Create a dictionary with the panel data
panel_data = {
    'Company': ['ABC', 'ABC', 'DEF', 'DEF', 'GHI', 'GHI'],
    'Year': [2001, 2002, 2001, 2002, 2001, 2002],
    'Revenue': [1234, 2345, 3456, 4567, 5678, 6789],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female']
}

# Create a DataFrame from the panel data dictionary
my_df = pd.DataFrame(panel_data)

# Print the original DataFrame
print("Original 'my_df' DataFrame:")
print(my_df)

# Create a DataFrame for female data
my_df_f = my_df[my_df['Gender'] == 'Female']

# Print the DataFrame for female data
print("\nDataFrame for Female Data:")
print(my_df_f)

# Create a DataFrame for male data
my_df_m = my_df[my_df['Gender'] != 'Female']

# Print the DataFrame for male data
print("\nDataFrame for Male Data:")
print(my_df_m)

# Concatenate the female and male DataFrames vertically (axis=0)
my_df_fm = pd.concat([my_df_f, my_df_m], ignore_index=True)

# Print the concatenated DataFrame
print("\nConcatenated DataFrame (Female + Male):")
print(my_df_fm)

# Separate DataFrames for female and male gender data
my_df_f_gender = my_df_f[['Gender', 'Revenue']]
my_df_m_gender = my_df_m[['Gender', 'Revenue']]

# Merge the female and male DataFrames based on the 'Gender' column using a left join
my_df_merged_left = pd.merge(my_df_f_gender, my_df_m_gender, on='Gender', how='left')

# Print the merged DataFrame (left join)
print("\nMerged DataFrame (Left Join) based on 'Gender' column:")
print(my_df_merged_left)

```
